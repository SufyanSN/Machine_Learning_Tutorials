```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3]
y = [1.2, 1.9 , 3.2]

plt.plot(x,y,'bo');
```

```python
import math
```

```python
''' Y = m * X + c  i.e Mean '''

sigma = 1
prob = []

for theta in range(0,361,1):
    for c in range(0,101,1):
        local_prob_i = []
        for i in range(0,3,1):

            Y = (math.tan(theta) * x[i]) + c
            prob_i = ( 1/(np.sqrt(2*np.pi)*sigma) ) * np.exp( -((Y - y[i])**2/2*(sigma**2)) )

            local_prob_i.append(prob_i)

            if theta == 0 and c == 0:
                prob.append(prob_i)

            elif i >= 2 and np.log(local_prob_i[0]*local_prob_i[1]*local_prob_i[2]) > np.log(prob[0]*prob[1]*prob[2]):
                prob[0] = local_prob_i[0]
                prob[1] = local_prob_i[1]
                prob[2] = local_prob_i[2]

                theta_max = theta
                c_max = c
```

```python
print(theta_max,c_max)
```

```python
x1 = [1, 2, 3]
y1 = []

for pt_x in x1:
    y1.append((math.tan(theta_max) * pt_x) + c_max)

print(x1)
print(y1)

plt.plot(x1,y1,'ro');
plt.plot(x1,y1,'r');
```
